```python
""" Mann-Whitney U Test with made up data comparing test scores of two independent groups:
    students who used two different study methods.
"""

#Group A: Traditional Study Method
group_a = [85, 90, 78, 92, 88, 76, 95, 89, 84, 91]

#Group B: New Study App
group_b = [82, 87, 90, 85, 88, 91, 84, 89, 86, 83]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mannwhitneyu

#Data
group_a = [85, 90, 78, 92, 88, 76, 95, 89, 84, 91]
group_b = [82, 87, 90, 85, 88, 91, 84, 89, 86, 83]

#mann-whitney U test
stat, p = mannwhitneyu(group_a, group_b, alternative='two-sided')
print(f'Mann-Whitney U Test Statistic: {stat}, p-value: {p}')

#visualization
sns.boxplot(data=[group_a, group_b])
plt.xticks([0, 1], ['Traditional Study Method', 'New Study App'])
plt.ylabel('Test Scores')
plt.title('Comparison of Test Scores by Study Method')
plt.show()

```

```python

# Interpretation of results 
if p < 0.05:
    print("There is a statistically significant difference between the two groups.")
else:
    print("There is no statistically significant difference between the two groups.")   
    
# Note: The p-value threshold of 0.05 is commonly used to determine statistical significance.

   
```

```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Add shared utils to sys.path
sys.path.append(str(Path().resolve().parents[2] / "shared" / "utils"))

from data_loading import load_csv

# Load data
df = load_csv('netflix_titles.csv')
if df is not None:
    display(df.head())
    df.info()
else:
    print("❌ Failed to load your_data.csv")
    
# # --- Data loading ---
# DATA_FILE = "netflix_titles.csv"
# df = load_csv(DATA_FILE)
# if df is not None:
#     display(df.head())
#     df.info()
# else:
#     print(f"❌ Failed to load {DATA_FILE}")
```

```python
# dataset overview
if df is not None:
    print("Shape:", df.shape)
    print("\nColumns:", df.columns.tolist())
    display(df.describe(include='all').T)
    print("\nMissing values:\n", df.isnull().sum())
    
if df is not None:
    print("Rows:", len(df))
    print("Columns:", len(df.columns))
```

```python
# Data Types and Unique Values
# Add a summary of data types and unique values per column:
if df is not None:
    print("\nData types:")
    print(df.dtypes)
    print("\nUnique values per column:")
    print(df.nunique())
```

```python
# duplicate rows  
if df is not None:
    num_duplicates = df.duplicated().sum()
    print(f"\nNumber of duplicate rows: {num_duplicates}")
```

```python
# Missing values
if df is not None:
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    if not missing.empty:
        print("Columns with missing values:")
        print(missing)
    else:
        print("No missing values.")  
```

```python
# Numeric features distribution. 
if df is not None:
    num_cols = df.select_dtypes(include=[np.number]).columns
    if len(num_cols) > 0:
        df[num_cols].hist(figsize=(12, 8), bins=30)
        plt.suptitle("Numeric Feature Distributions")
        plt.show()
    else:
        print("No numeric columns found.")
```

```python
# Outlier Detection (for numeric columns)
if df is not None and 'duration_num' in df.columns:
    q1 = df['duration_num'].quantile(0.25)
    q3 = df['duration_num'].quantile(0.75)
    iqr = q3 - q1
    outliers = df[(df['duration_num'] < (q1 - 1.5 * iqr)) | (df['duration_num'] > (q3 + 1.5 * iqr))]
    print(f"\nNumber of outliers in duration_num: {len(outliers)}")
```

```python
# categorical features distribution
if df is not None:
    cat_cols = df.select_dtypes(include=['object', 'category']).columns
    for col in cat_cols:
        print(f"\nValue counts for {col}:")
        print(df[col].value_counts(dropna=False).head(10))
        if df[col].nunique() < 20:
            sns.countplot(y=col, data=df, order=df[col].value_counts().index)
            plt.title(f"Distribution of {col}")
            plt.show()
```

```python
# # Correlations 
# if df is not None and len(num_cols) > 1:
#     corr = df[num_cols].corr()
#     plt.figure(figsize=(10, 8))
#     sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
#     plt.title("Correlation Matrix")
#     plt.show()


if df is not None and len(df.select_dtypes(include=[np.number]).columns) > 1:
    num_cols = df.select_dtypes(include=[np.number]).columns
    corr = df[num_cols].corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
    plt.title("Correlation Matrix")
    plt.show()
```

```python
### Gives you a full overview, missing values, and key distributions.
### Visualizes content type, country, year, genre, duration, and seasons.
### Prints out top-level insights at the end.

# Content Type Distribution
if df is not None and 'type' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x='type')
    plt.title('Content Type Distribution')
    plt.show()
    print(df['type'].value_counts())

# Top Countries by Content Count
if df is not None and 'country' in df.columns:
    country_series = df['country'].dropna().str.split(',', expand=True).stack().str.strip()
    top_countries = country_series.value_counts().head(10)
    plt.figure(figsize=(8, 4))
    sns.barplot(y=top_countries.index, x=top_countries.values, orient='h')
    plt.title('Top 10 Countries by Content Count')
    plt.xlabel('Count')
    plt.ylabel('Country')
    plt.show()
    print(top_countries)

# Content Added Over Time
if df is not None and 'date_added' in df.columns:
    df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
    added_per_year = df['date_added'].dt.year.value_counts().sort_index()
    plt.figure(figsize=(10, 4))
    sns.lineplot(x=added_per_year.index, y=added_per_year.values, marker='o')
    plt.title('Content Added per Year')
    plt.xlabel('Year')
    plt.ylabel('Number of Titles Added')
    plt.show()
    print(added_per_year)
    
if df is not None and 'date_added' in df.columns:
    df['month_added'] = df['date_added'].dt.month
    plt.figure(figsize=(10, 4))
    sns.countplot(x='month_added', data=df)
    plt.title('Content Added by Month')
    plt.xlabel('Month')
    plt.ylabel('Number of Titles Added')
    plt.show()

# Most Common Genres
if df is not None and 'listed_in' in df.columns:
    genre_series = df['listed_in'].dropna().str.split(',', expand=True).stack().str.strip()
    top_genres = genre_series.value_counts().head(10)
    plt.figure(figsize=(8, 4))
    sns.barplot(y=top_genres.index, x=top_genres.values, orient='h')
    plt.title('Top 10 Genres')
    plt.xlabel('Count')
    plt.ylabel('Genre')
    plt.show()
    print(top_genres)

# Movie Duration Distribution (if available)
if df is not None and 'duration' in df.columns:
    # Extract numeric duration for movies
    df['duration_num'] = df['duration'].str.extract(r'(\d+)').astype(float)
    plt.figure(figsize=(8, 4))
    sns.histplot(df[df['type'] == 'Movie']['duration_num'].dropna(), bins=30, kde=True)
    plt.title('Movie Duration Distribution (minutes)')
    plt.xlabel('Duration (min)')
    plt.show()
    print(df[df['type'] == 'Movie']['duration_num'].describe())

# TV Show Seasons Distribution (if available)
if df is not None and 'duration' in df.columns:
    df['seasons_num'] = df['duration'].str.extract(r'(\d+)').astype(float)
    plt.figure(figsize=(8, 4))
    sns.countplot(x='seasons_num', data=df[df['type'] == 'TV Show'])
    plt.title('TV Show Seasons Distribution')
    plt.xlabel('Number of Seasons')
    plt.show()
    print(df[df['type'] == 'TV Show']['seasons_num'].value_counts().sort_index())

# --- Example Insights (print or markdown) ---
print("\n--- Insights ---")
if df is not None:
    print(f"Most content is: {df['type'].value_counts().idxmax()}")
    print(f"Top country for content: {top_countries.idxmax()}")
    print(f"Most common genre: {top_genres.idxmax()}")
    print(f"Year with most content added: {added_per_year.idxmax()}")
```

```python
if df is not None and 'director' in df.columns:
    print("\nTop 10 directors:")
    print(df['director'].value_counts().head(10))

if df is not None and 'cast' in df.columns:
    cast_series = df['cast'].dropna().str.split(',', expand=True).stack().str.strip()
    print("\nTop 10 actors:")
    print(cast_series.value_counts().head(10))

if df is not None and 'rating' in df.columns:
    print("\nContent rating distribution:")
    print(df['rating'].value_counts())
    plt.figure(figsize=(8, 4))
    sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index)
    plt.title('Content Rating Distribution')
    plt.show()
```
